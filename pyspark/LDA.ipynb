```python
docTopic = sc.textFile("./python/docTopic")
list1 = []
for i in range(8) :
    list1.append(StructField(str(i), FloatType(), True))
schema = StructType(list1)
def f(x):
    list1 = []
    words = x.split(",")
    for i in range(len(words)) :
        list1.append(float(words[i].strip().strip(']').strip('[')))
    return list1
pandas_docTopic = spark.createDataFrame(docTopic.map(f).collect(), schema).toPandas()
```

```python
topicWord = sc.textFile("./python/topicWord")
list1 = []
for i in range(8) :
    list1.append(StructField(str(i), FloatType(), True))
schema = StructType(list1)
def f(x):
    list1 = []
    words = x.split(",")
    for i in range(len(words)) :
        list1.append(float(words[i].strip().strip(']').strip('[')))
    return list1
pandas_topicWord = spark.createDataFrame(topicWord.map(f).collect(), schema).toPandas()
```
